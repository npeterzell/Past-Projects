# Data Cleaning

Seven datasets were used in total, one containing the population and total income of each US county, and 6 containing the vaccination distribution data of each state. Each dataset was formatted differnetly and thus required different cleaning methods however the end goal was to produce a combined dataset where each row represented a unique county and the columns contained the income per capita and vaccines administered per capita of that county.

The census data had to be grouped according to a column as the rows containing income information were split to mulitples of 3 rows. This dataset also had missing values and duplicate counties which needed to be addressed. A function was applied to the dataset to compute the income per capita column of the final dataset.

As we were unable to locate a national dataset with per county granularity we opted to use individual state's datasets which did report per county. This made the data cleaning process challenging as each state had a slightly different reporting format. As there were multiple counties with the same name in the US a state column was nesassary to create before preforming the merge to the main dataset which would ensure accurate data points and non duplication.



```python
income_raw = pd.read_csv("Datasets/IncomeByCounty.csv")
vaccine_count_cal = pd.read_csv("Datasets/DosesAdminsteredbyCounty(California).csv", thousands=',')
vaccine_count_wash = pd.read_csv("Datasets/Doses(Washington).csv", thousands=',')
vaccine_count_geor = pd.read_csv("Datasets/Georgia Doses.csv", thousands=',')
vaccine_count_ohio = pd.read_csv("Datasets/Ohio Doses.csv", thousands=',')
vaccine_count_ariz = pd.read_csv("Datasets/Arizona Doses.csv", thousands=',')
vaccine_count_kans = pd.read_csv("Datasets/Kansas_Vaccine.csv")

group = income_raw.groupby("LineCode")

counties = group.get_group(1)["GeoName"].reset_index()
del counties["index"]


income_data = group.get_group(1)["2019"].reset_index()
del income_data["index"]

population_data = group.get_group(2)["2019"].reset_index()
del population_data["index"]

county_data = pd.DataFrame(counties)
county_data = county_data.rename(columns={"GeoName": "County"})

county_data["Total Income"] = income_data
county_data["Population"] = population_data


#Remove NA from Dataset for accurate int datatype
county_data = county_data.replace("(NA)", "1")
#Convert object data to int
county_data["Total Income"] = pd.to_numeric(county_data["Total Income"])
county_data["Population"] = pd.to_numeric(county_data["Population"])


#Calculate Income per Capita
percapitaincome = county_data.apply(lambda row: row["Total Income"] / row["Population"], axis = 1).reset_index()
del percapitaincome["index"]
county_data["Income per Capita"] = percapitaincome
county_data.to_csv("CleanedIncomePerCapita.csv")

#Seperate County and State
count_state_temp = county_data["County"].str.split(", ", n = 1, expand = True) 
county_data.drop(columns =["County"], inplace = True) 

county_data["County"] = count_state_temp[0]
county_data["State"] = count_state_temp[1]

#filter only california counties
cal_income = county_data[(county_data.State == "CA")]

#filter only washington counties
wash_income = county_data[(county_data.State == "WA")]

#filter only arizona counties
ariz_income = county_data[(county_data.State == "AZ") | ((county_data.State == "AZ*"))]

#filter only ohio counties
ohio_income = county_data[(county_data.State == "OH")]

#filter only georgia counties
geor_income = county_data[(county_data.State == "GA")]

#filer only kansas counties
kans_income = county_data[(county_data.State == "KS")]
```

# Data Analysis & Results

We first created dictionaries and functions that would aid in our analysis of the data. See function header for description of functionality. We also calculated doses per capita as a metric to control for the effect of population size.

```python
#define dictionaries for mapping
color_dict = {"all": "black", "cal": "blue", "wash": "red", "geor": "pink", "ohio": "green", "ariz": "yellow", "kans": "purple"}
state_name = {"all": "All 6 states", "cal": "California", "wash": "Washigton", "ariz": "Arizona", "ohio": "Ohio", "geor": "Georgia", "kans" : "Kansas"}
dataset_dict = {"all": all_states,"cal": merged_cal, "wash": merged_wash, "ariz": merged_ariz, "ohio": merged_ohio, "geor": merged_geor, "kans":merged_kans}
```

```python
#Calculate doses per capita for each state
for state in states:
    vaccine_per_capita = dataset_dict[state].apply(lambda row: row["Doses Administered"] / row["Population"], axis = 1)
    dataset_dict[state]["Doses per Capita"] = vaccine_per_capita
```

## Analysis Functions 

```python
num_bins = 25
def plot_hist(state):
    """
    Function to plot histograms of Income per Capita, Doses Administered and Doses per Capita.
    Also provides mean and correlation information

    Parameters
    ----------
    state: state to be plotted
    
    """
    f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))
    print("Income per Capita Centered Around: " + str(dataset_dict[state]["Income per Capita"].mean()) + " thousand dollars")
    print("Doses Administered Centered Around: " + str(dataset_dict[state]["Doses Administered"].mean()) + " Doses")
    print("Doses per Capita Centered Around: " + str(dataset_dict[state]["Doses per Capita"].mean()) + " Doses")
    print("Correlation between Income per Capita and Doses per Capita: " + str(dataset_dict[state]["Income per Capita"].corr(dataset_dict[state]["Doses per Capita"])))
    ax1.hist(dataset_dict[state]["Income per Capita"], num_bins, facecolor=color_dict[state], alpha=0.7)
    ax1.set_title("Income per Capita " + state_name[state])
    ax1.set_xlabel("Income per Capita")
    ax2.hist(dataset_dict[state]["Doses Administered"], num_bins, facecolor=color_dict[state], alpha=0.7)
    ax2.set_title("Doses Administered " + state_name[state])
    ax2.set_xlabel("Doses Administered ")
    ax3.hist(dataset_dict[state]["Doses per Capita"], num_bins, facecolor=color_dict[state], alpha=0.7)
    ax3.set_title("Doses per Capita " + state_name[state])
    ax3.set_xlabel("Doses per Capita")
    plt.show()
```

```python
def plot_scatter_ipc_vs_da(state):
    """
    Function to plot scatter plot of Income per Capita vs Doses Administered

    Parameters
    ----------
    state: state to be plotted
    ax: axes to be plotted on
    """
    if (state != "all"):
        plt.plot(dataset_dict[state]["Income per Capita"], dataset_dict[state]["Doses Administered"], color = color_dict[state], marker='o',
                 alpha = 0.5, linestyle="None", label = state_name[state])
    else:
        for state in state_list:
            plt.plot(dataset_dict[state]["Income per Capita"], dataset_dict[state]["Doses Administered"], color = color_dict[state], marker='o',
             alpha = 0.5, linestyle="None", label = state_name[state])
            plt.legend()
    plt.xlabel("Income per Capita (in Thousands)")
    plt.ylabel("Doses Administered")
    plt.title("Income per Capita vs Doses Administered" )
```

```python
def plot_scatter_ipc_vs_dpc(state, ax):
    """
    Function to plot scatter plot of Income per Capita vs Doses per Capita
    Must specify axes to plot on

    Parameters
    ----------
    state: state to be plotted
    ax: axes to plot on
    """
    if (state != "all"):
        ax.plot(dataset_dict[state]["Income per Capita"], dataset_dict[state]["Doses per Capita"], color = color_dict[state], marker='o',
                 alpha = 0.5, linestyle="None", label = state_name[state])
    else:
        for state in state_list:
            ax.plot(dataset_dict[state]["Income per Capita"], dataset_dict[state]["Doses per Capita"], color = color_dict[state], marker='o',
             alpha = 0.5, linestyle="None", label = state_name[state])
        ax.legend()
    ax.set_xlabel("Income per Capita (in Thousands)")
    ax.set_ylabel("Doses per Capita")
    ax.set_title("Income per Capita vs Doses per Capita")
```

```python
def plot_scatter_dpc_vs_da(state):
    """
    Function to plot scatter plot of Doses per Capita vs Doses Administered

    Parameters
    ----------
    state: state to be plotted
    ax: axes to be plotted on
    """
    if (state != "all"):
        plt.plot(dataset_dict[state]["Doses per Capita"], dataset_dict[state]["Doses Administered"], color = color_dict[state], marker='o',
                 alpha = 0.5, linestyle="None", label = state_name[state])
    else:
        for state in state_list:
            plt.plot(dataset_dict[state]["Doses per Capita"], dataset_dict[state]["Doses Administered"], color = color_dict[state], marker='o',
             alpha = 0.5, linestyle="None", label = state_name[state])
            plt.legend()
    plt.xlabel("Doses per Capita")
    plt.ylabel("Doses Administered")
    plt.title("Doses per Capita vs Doses Administered" )
```

```python
def linear_model_state(state):
    """
    Function to use linear regression to model relationship between Incompe per Capita and Doses per Capita
    Does not remove outlier
    Plots model

    Parameters
    ----------
    state: state to be plotted
    
    """
    state_lin = dataset_dict[state].rename(columns = {"Income per Capita": "IPC", "Doses per Capita": "DPC"})
    state_lin.head()
    outcome_s, predictors_s = patsy.dmatrices(formula_like = "DPC ~ IPC", data = state_lin)
    mod_s = sm.OLS(outcome_s, predictors_s)
    res_s = mod_s.fit()
    print(res_s.summary())
    fig, ax = plt.subplots()
    plot_scatter_ipc_vs_dpc(state, ax)
    xs = np.arange(dataset_dict[state]['Income per Capita'].min(), dataset_dict[state]['Income per Capita'].max())
    ys = res_s.params[0] + (res_s.params[1] * xs)
    ax.plot(xs, ys, '--k', linewidth=4, label='Model')
    ax.legend()
    ax.set_title("Linear model of Income per Capita vs Doses per Capita")
```

```python
def remove_outlier(df):
    """
    Fuction to remove outliers from dataframe
    
    Parameters
    ----------
    df: dataframe to remove outliers from
    """
    rem_out = df[(df["Income per Capita"] < INCOME_CAP) & (df["Doses per Capita"] < DOSES_MAX) & (df["Doses per Capita"] > DOSES_MIN)]
    return rem_out
```

```python
def outlier(df):
    """
    Function to show outliers
    """
    out = df[(df["Income per Capita"] >= INCOME_CAP) | (df["Doses per Capita"] >= DOSES_MAX) | (df["Doses per Capita"] <= DOSES_MIN)]
    return out
```

```python
def plot_ipc_vs_dpc_wo_out(state, ax):
    """
    Function to plot scatter plot of Income per Capita vs Doses per Capita without outliers
    Removes outlier
    Must specify axes to plot on

    Parameters
    ----------
    state: state to be plotted
    ax: axes to plot on
    """
    if (state != "all"):
        state_wo_out = remove_outlier(dataset_dict[state])
        ax.plot(state_wo_out["Income per Capita"], state_wo_out["Doses per Capita"], color = color_dict[state], marker='o',
                 alpha = 0.5, linestyle="None", label = state_name[state])
    else:
        for state in state_list:
            state_wo_out = remove_outlier(dataset_dict[state])
            ax.plot(state_wo_out["Income per Capita"], state_wo_out["Doses per Capita"], color = color_dict[state], marker='o',
             alpha = 0.5, linestyle="None", label = state_name[state])
        ax.legend()
    ax.set_xlabel("Income per Capita (in Thousands)")
    ax.set_ylabel("Doses per Capita")
    ax.set_title("Income per Capita vs Doses per Capita [Outliers Removed]")
```

```python
def linear_model_state_rem_out(state):
    """
    Function to use linear regression to model relationship between Incompe per Capita and Doses per Capita without outliers
    Removes Outliers
    Plots model

    Parameters
    ----------
    state: state to be plotted
    """
    state_df = dataset_dict[state]
    state_rem_out = remove_outlier(state_df)
    state_lin = state_rem_out.rename(columns = {"Income per Capita": "IPC", "Doses per Capita": "DPC"})

    outcome_s, predictors_s = patsy.dmatrices(formula_like = "DPC ~ IPC", data = state_lin)
    mod_s = sm.OLS(outcome_s, predictors_s)
    res_s = mod_s.fit()
    print(res_s.summary())
    print("Correlation between Income per Capita and Doses per Capita: " + str(state_rem_out["Income per Capita"].corr(state_rem_out["Doses per Capita"])))
    
    fig, ax = plt.subplots()
    plot_ipc_vs_dpc_wo_out(state, ax)
    
    xs = np.arange(state_rem_out['Income per Capita'].min(), state_rem_out['Income per Capita'].max())
    ys = res_s.params[0] + (res_s.params[1] * xs)
    ax.plot(xs, ys, '--k', linewidth=4, label='Model')
    ax.legend()
```

### Removing Outliers
